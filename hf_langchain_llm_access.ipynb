## Accessing Open-Source or Closed-Source LLMs as Endpoints

```python
# To access a Hugging Faceâ€“deployed LLM (open-source or closed-source) as an API endpoint, we use HuggingFaceEndpoint.
# This allows LangChain to interact with the model without loading it locally, reducing RAM usage.
# When accessed via an endpoint, LangChain treats the model as a regular (non-chat) LLM.
# Currently, there is no direct chat-native interface for Hugging Face endpoints in LangChain.
# Therefore, all endpoint-based models must first be accessed as regular LLMs, and then optionally adapted for chat-style usage if needed.

# Note: Using endpoints is the only practical way to access deployed Hugging Face models remotely, even though they are handled as regular LLMs in LangChain.
```

```python
model = HuggingFaceEndpoint(repo_id="deepseek-ai/DeepSeek-V3.2" ,temperature =0.5)
```

```python
model.invoke("what is the capital of india") #It is Chatmodel
```

```python
# Note: This error occurs because the Gemma instruction-tuned model only supports the conversational task, but LangChain tries to use it as a text-generation model by default when accessed through an endpoint.
```

```python
#Why gives an error?
# Whenever we are accessing models through an Endpoint, LangChain will assume it is a regular model.
# So to solve this problem we just wrapped the model into ChatHuggingFace.
```

```python
cm = ChatHuggingFace(llm=model)
```

```python
response = cm.invoke("what is the capital of india")
```

```python
response.content
```
